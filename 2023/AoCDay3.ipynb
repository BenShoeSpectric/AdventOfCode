```python
from pathlib import Path
import re

path = Path('Day3Part1.txt')
contents = path.read_text().splitlines()
engineSize = len(contents)
lineLength = len(contents[0])
```

```python
def adder(part,engineLine,lineAbove,lineBelow):
    
    partLen = len(part)
    partLoc = engineLine.find(part)
    t = ""
#    print("Part:",part,"Loc:",partLoc,"Len:",partLen)

    try:
        End = engineLine[(partLoc+partLen)]
    except IndexError:
        End = ""

    if partLoc == 0:
        Begining = "."
        Start = 0
    else:
        Begining = engineLine[(partLoc-1)]
        Start = partLoc - 1
    
    s = [lineAbove[Start:(partLoc+partLen+1)],lineBelow[Start:(partLoc+partLen+1)],Begining,End]
#            print(Begining, End)
    Surround = t.join(s)
#            print("Surrounding Line:",Surround)
    Surround = list(filter(None, Surround.split(".")))
#            print(Surround, len(Surround))
    if len(Surround) > 0:
       # print(part)
        return int(part)
    else:
        return 0
```

```python
Total = 0
for i in range(0,engineSize):
    engineLine = contents[i]
#    print(engineLine)
    
    if i != 0:
        lineAbove = contents[i-1]
#        print("Line Above:",lineAbove)
    else:
        lineAbove = ""
#        print("Line Above:",lineAbove)
    
    if i != (engineSize-1):
        lineBelow = contents[i+1]
#        print("Line Below:",lineBelow)
    else:
        lineBelow = ""
#        print("Line Below:",lineBelow)

    engineList = list(filter(None,re.split('[^0-9]', engineLine)))
    
    ###  Correct until here ###
    #print("engineList:", engineList)
    for part in engineList:
        partValue = adder(part,engineLine,lineAbove,lineBelow)

        Total += partValue
        print("Running Total:",Total,"Value:",partValue)

print("Total of Schematic is:", Total)
```
